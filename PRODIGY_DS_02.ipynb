```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

csv_path = r"C:\Users\Saneeka\Downloads\train.csv"
df = pd.read_csv(csv_path)

df.isnull().sum()

df['Age'].fillna(df['Age'].median(), inplace=True)

df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

df.drop(columns=['Cabin'], inplace=True)

df.dropna(subset=['Fare'], inplace=True)

df.isnull().sum()
```

```python
df.describe()

sns.countplot(data=df, x='Survived')
plt.title('Distribution of Survived')
plt.show()
```

```python
sns.countplot(data=df, x='Sex')
plt.title('Distribution of Sex')
plt.show()
```

```python
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Distribution of Age')
plt.show()
```

```python
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title('Survival Rate by Sex')
plt.show()
```

```python
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Survived'] == 1]['Age'].dropna(), bins=30, kde=True, color='teal', label='Survived')
sns.histplot(df[df['Survived'] == 0]['Age'].dropna(), bins=30, kde=True, color='yellow', label='Not Survived')
plt.legend()
plt.title('Age Distribution by Survival Status')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Survived'] == 1]['Fare'].dropna(), bins=30, kde=True, color='black', label='Survived')
sns.histplot(df[df['Survived'] == 0]['Fare'].dropna(), bins=30, kde=True, color='teal', label='Not Survived')
plt.legend()
plt.title('Fare Distribution by Survival Status')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
